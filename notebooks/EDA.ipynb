```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../vehicles_us.csv')
```

```python
display(df.describe())
display(df.shape)
display(df.info())
display(df.head())
```

```python
# Half of the is_4wd column is missing. The best course of action here would probably be dropping it as 50%
# is a substancial amount of data missing. 
# Model_year, cylinders, and odometer should be an int while date_posted should be a datetime object.
# Although, these 3 columns have NaN which is a float type, so we may have to replace this value with the
# mean of each column to clean this up. Will leave missing paint color values as is. 
display(df.isna().sum())
df = df.drop('is_4wd', axis=1)
df['model_year'] = df['model_year'].where(df['model_year'].isna() == False, df['model_year'].mean())
df['cylinders'] = df['cylinders'].where(df['cylinders'].isna() == False, df['cylinders'].mean())
df['odometer'] = df['odometer'].where(df['odometer'].isna() == False, df['odometer'].mean())
df[['model_year', 'cylinders', 'odometer']] = df[['model_year', 'cylinders', 'odometer']].astype('int')
df['date_posted'] = pd.to_datetime(df['date_posted'], format='%Y-%m-%d')
display(df.isna().sum())
display(df.head())
df.info()
```
